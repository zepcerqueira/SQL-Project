```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
web_scrape_comments = pd.read_csv("web_scrape_comments.csv")
```

## Using VADER

### VADER (Valence Aware Dictionary and sEntiment Reasoner) is a rule-based sentiment analyzer that has been trained on social media text.

```python
def get_sentiment_scores(comment):
    sia = SentimentIntensityAnalyzer()
    scores = sia.polarity_scores(comment)
    return pd.Series(scores)

web_scrape_comments[["neg", "neu", "pos", "compound"]] = web_scrape_comments["review_body"].apply(get_sentiment_scores)
```

```python
columns_to_keep = ["id", "name", "jobTitle", "tagline", "comments", "overall", "curriculum", "jobSupport", "review_body", "school", "neg", "neu", "pos", "compound"]

web_scrape_comments_filtered = web_scrape_comments[columns_to_keep]
```

```python
columns = ["overall", "school", "neg", "neu", "pos", "compound"]

sns.pairplot(web_scrape_comments_filtered[columns])
plt.show()
```
