```python
import pandas as pd
from simpletransformers.classification import ClassificationModel
```

```python
def get_sentiment_scores(comment):
    model = ClassificationModel("roberta", "roberta-base", use_cuda=False)  # You can choose a different model if needed
    result, _ = model.predict([comment])
    
    # Normalize the sentiment score to the range of 0 to 5
    normalized_score = (result[0] + 1) * 2.5
    
    return normalized_score

# Apply the function to each comment in the DataFrame and create a new column 'sentiment_score'
web_scrape_comments['sentiment_score'] = web_scrape_comments['review_body'].apply(get_sentiment_scores)

# Display the DataFrame with the new column
print(web_scrape_comments)
```

```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer

# Assuming you have a DataFrame with a 'comments' column
# web_scrape_comments = pd.DataFrame({'comments': ["This is great!", "Not impressed.", "I love it!", "It's okay."]})

# Load the VADER sentiment analysis tool
sia = SentimentIntensityAnalyzer()

# Define a function to convert sentiment scores to ratings
def score_to_rating(sentiment_score):
    if sentiment_score >= 0.6:
        return 5
    elif 0.2 <= sentiment_score < 0.6:
        return 4
    elif -0.2 <= sentiment_score < 0.2:
        return 3
    elif -0.6 <= sentiment_score < -0.2:
        return 2
    else:
        return 1

# Apply sentiment analysis and map scores to ratings
web_scrape_comments['sentiment_score'] = web_scrape_comments['review_body'].apply(lambda x: sia.polarity_scores(x)['compound'])
web_scrape_comments['rating'] = web_scrape_comments['sentiment_score'].apply(score_to_rating)

# Display the DataFrame with sentiment scores and ratings
web_scrape_comments
```
